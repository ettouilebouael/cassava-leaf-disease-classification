```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import torch
from sklearn.model_selection import train_test_split
```

## Visualisation des images

```python
train_path = "./data/train.csv"
imgs_path = "./data/train_images/"
```

```python
df = pd.read_csv(train_path)
```

```python
df.head()
```

```python
df.label.plot(kind="hist")
```

```python
#load one image
img_id = 1000
img_name , label = df.values[img_id]
image_path = f"{imgs_path}{img_name}"
img = plt.imread(image_path)
plt.imshow(img)
plt.ylabel(label)
```

```python
# Visualize multiple images
img_ids =  np.random.randint(0,len(df),3*4) 
fig, axs = plt.subplots(3, 4, figsize=(30,10))
fig.suptitle('Train images')
for ax, img_id in zip(axs.flatten(), img_ids):
    img_name , label = df.values[img_id]
    image_path = f"{imgs_path}{img_name}"
    img = plt.imread(image_path)
    ax.imshow(img)
    ax.set_title(f'Label : {label}')
    plt.plot()
```
